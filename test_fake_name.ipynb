```python
import datasets
import numpy as np
from obfuscation.fake_name import FakeName
```

```python
wfu_dataset = datasets.load_dataset('wfudata', trust_remote_code=True)
fakename = FakeName('obfuscation/fake_name_list.txt')
```

```python
for i in range(len(wfu_dataset['train'])):
    phi = wfu_dataset['train']['phi'][i]
    text = wfu_dataset['train']['text'][i]
    names = []
    for phi_type, offset in zip(phi['types'], phi['offsets']):
        if phi_type == 'NAME':
            names.append(text[offset[0]:offset[1]])
    print(f'----- {names} ------ ')
    print(fakename.parse_and_replace(names))
```
