```python
import datasets
import numpy as np
```

```python
wfu_dataset = datasets.load_dataset('wfudata', trust_remote_code=True)
```

```python
all_names = []

for i in range(len(wfu_dataset['train'])):
    phi = wfu_dataset['train']['phi'][i]
    text = wfu_dataset['train']['text'][i]
    for phi_type, offset in zip(phi['types'], phi['offsets']):
        if phi_type == 'NAME':
            all_names.append(text[offset[0]:offset[1]])
```

```python
from collections import defaultdict
```

```python
part_counts = defaultdict(list)
for x in all_names:
    for y in x.split():
        y = y.lower().strip()
        if len(y) > 1:
            part_counts[y].append(x)
```

```python
credentials = ['dr', 'drs', 'ms', 'md', 'phd', 'mr', 'jr', 'sr', 'pa', 'do']
```

```python
sorted(part_counts.items(), key=lambda x:len(x[1]), reverse=True)
```

```python
part_counts = defaultdict(list)
for x in all_names:
    for y in x.split():
        y = y.lower().strip()
        if y.find('.') >= 0 and len(y) < 10 and len(y) > 2:
            part_counts[y].append(x)
```

```python
sorted(part_counts.items(), key=lambda x:len(x[1]), reverse=True)
```

```python
cc = set()
with open('credential_list.txt', 'r') as fid:
    for line in fid:
        cc.add(line.split()[0].lower())
```

```python
for x in all_names:
    for y in x.split():
        y = y.lower().strip().replace('.','')
        if y in cc and y not in credentials:
            print(y, '---', x)
```

```python
import re
```

```python
pat1 = re.compile(r'[ |,]')
```

```python
for name in all_names:
    prev_end = 0
    for span in pat1.finditer(name):
        curr_start, curr_end = span.span(0)
        curr_word = name[prev_end:curr_start]
        if curr_word.lower().replace('.','') in credentials:
            print(name)
        prev_end = curr_end
```
