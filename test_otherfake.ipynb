```python
import datasets
from datetime import datetime
import time
```

```python
wfu_dataset = datasets.load_dataset('wfudata', trust_remote_code=True)
```

```python
all_hospitals, all_locations, all_others = [], [], []

for i in range(len(wfu_dataset['train'])):
    phi = wfu_dataset['train']['phi'][i]
    text = wfu_dataset['train']['text'][i]
    prev_offset = None
    for phi_type, offset in zip(phi['types'], phi['offsets']):
        if phi_type == 'HOSPITAL':
            all_hospitals.append(text[offset[0]:offset[1]])
        if phi_type == 'LOCATION':
            all_locations.append(text[offset[0]:offset[1]])
        if phi_type == 'OTHER':
            all_others.append(text[offset[0]:offset[1]])
```

```python
len(set([x.strip() for x in all_hospitals])), len(set([x.strip() for x in all_locations])), len(set([x.strip() for x in all_others]))
```
