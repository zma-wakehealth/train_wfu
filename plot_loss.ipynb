```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def extract_loss(filename):
    training_loss = []
    eval_loss = []
    name_f1 = []

    with open(filename, 'r') as fid:
        for line in fid:
            if line.startswith('training'):
                training_loss.append(float(line.split('=')[-1]))
            elif line.startswith('weighted avg'):
                eval_loss.append(float(line.split()[4]))
            elif line.lstrip().startswith('NAME'):
                name_f1.append(float(line.split()[3]))

    return training_loss, eval_loss, name_f1
```

```python
training_loss_1, eval_loss_1, name_f1_1 = extract_loss('no_balanced.txt')
training_loss_2, eval_loss_2, name_f1_2 = extract_loss('no_balanced_biodistilbert.txt')
```

```python
plt.plot(training_loss_1, '.-');
plt.plot(training_loss_2, 'x')
plt.ylim([0,0.1])
```

```python
_, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].plot(name_f1_1, '.-');
axs[0].plot(name_f1_2, 'x');

axs[1].plot(eval_loss_1, '.-');
axs[1].plot(eval_loss_2, 'x');

plt.ylim([0.98, 1.0])
```

```python
training_loss_1, eval_loss_1, name_f1_1 = extract_loss('with_balanced_weighting.txt')
training_loss_2, eval_loss_2, name_f1_2 = extract_loss('with_balanced_weighting_biodistilbert.txt')
```

```python
plt.plot(training_loss_1, '.-');
plt.plot(training_loss_2, 'x')
```

```python
_, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].plot(name_f1_1, '.-');
axs[0].plot(name_f1_2, 'x');

axs[1].plot(eval_loss_1, '.-');
axs[1].plot(eval_loss_2, 'x');

plt.ylim([0.98, 1.0])
```
