```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def extract_loss(filename):
    training_loss = []
    eval_loss = []
    name_f1 = []

    with open(filename, 'r') as fid:
        for line in fid:
            if line.startswith('training'):
                training_loss.append(float(line.split('=')[-1]))
            elif line.startswith('weighted avg'):
                eval_loss.append(float(line.split()[4]))
            elif line.lstrip().startswith('NAME'):
                name_f1.append(float(line.split()[3]))

    return training_loss, eval_loss, name_f1
```

```python
training_loss, eval_loss, name_f1 = extract_loss('no_balanced.txt')
```

```python
plt.plot(training_loss, '.-');
```

```python
plt.plot(name_f1, 'o-');
plt.plot(eval_loss, 'rx-');
plt.ylim([0.98, 1.0])
```

```python
training_loss, eval_loss, name_f1 = extract_loss('with_balanced_weighting.txt')
```

```python
plt.plot(training_loss, '.-');
```

```python
plt.plot(name_f1, 'o-');
plt.plot(eval_loss, 'rx-');
plt.ylim([0.98, 1.0])
```
